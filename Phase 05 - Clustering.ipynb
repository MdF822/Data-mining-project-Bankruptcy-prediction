<div id="sec_intro" style="direction: rtl; line-height: 300%;">

        <p></p>
        با توجه به این که دیتای ما مقدار NULL ندارد، خودمان به صورت رندوم برخی خانه ها را NULL کردیم.

```python
# تعیین تعداد مقادیر null برای هر ستون
num_nulls = 10

# قراردادن مقادیر null به صورت تصادفی
for column in df.columns:
    # انتخاب 5 ایندکس تصادفی برای هر ستون
    random_indices = np.random.choice(df.index, size=num_nulls, replace=False)
    df.loc[random_indices, column] = np.nan
```

```python
# جایگزینی داده‌های گم‌شده با میانگین هر ستون
df_filled_mean = df.fillna(df.mean())

# # جایگزینی داده‌های گم‌شده با میانه هر ستون
# df_filled_median = df.fillna(df.median())
# df_filled_median
```

            <h3>حذف داده های پرت</h3>

    </font>

<div id="sec_intro" style="direction: rtl; line-height: 300%;">
    <font face="XB Zar" size="4">
        <p></p>
        با توجه به IQR، داده های پرت را شناسایی کرده و حذف کردیم.
    </font>
</div>

```python
# پردازش حذف مقادیر پرت و جایگزینی داده‌های گم‌شده برای هر ستون عددی
for column in df.select_dtypes(include=[np.number]).columns:
    # محاسبه حدود برای حذف مقادیر پرت
    q1 = df_filled_mean[column].quantile(0.25)
    q3 = df_filled_mean[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # حذف مقادیر پرت
    df_cleaned = df_filled_mean[(df_filled_mean[column] >= lower_bound) & (df_filled_mean[column] <= upper_bound)]

df_cleaned
```

<div id="sec_intro" style="direction: rtl; line-height: 300%;">
    <font face="XB Zar" size="4">
        <p></p>
        این کد با استفاده از تحلیل مؤلفه‌های اصلی (PCA) ابعاد داده‌ها را کاهش می‌دهد. در این فرایند،
واریانس توضیح داده‌شده، سهم هر مؤلفه اصلی در توضیح واریانس کل داده‌ها محاسبه می‌شود.
واریانس تجمعی، مجموع واریانس توضیح داده‌شده توسط مؤلفه‌ها بررسی می‌شود.
هدف، تعداد مؤلفه‌هایی که حداقل 95٪ از واریانس کل داده‌ها را توضیح می‌دهند، انتخاب می‌شود. این تضمین می‌کند که بیشترین اطلاعات با کمترین تعداد مؤلفه حفظ شود.
    </font>
</div>

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df_normalized)

# Step 2: Apply PCA
pca = PCA(n_components=None)  # If n_components is not specified, all components are calculated.
pca_result = pca.fit_transform(data_scaled)

# Step 3: Explained variance analysis
explained_variance = pca.explained_variance_ratio_  # Explained variance by each component
cumulative_variance = np.cumsum(explained_variance)  # Cumulative variance

# Plot cumulative variance
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.title('Cumulative Variance Plot')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Variance')
plt.grid()
plt.show()

# Step 4: Select the appropriate number of components (for example, 95% variance)
n_components = np.argmax(cumulative_variance >= 0.95) + 1
print(f"Number of components to explain 95% variance: {n_components}")

# Apply PCA with the selected number of components
pca_final = PCA(n_components=n_components)
data_reduced = pca_final.fit_transform(data_scaled)

# Display reduced data
print("Shape of reduced data:")
print(data_reduced.shape)
```

```python
# تبدیل داده‌های کاهش‌یافته به DataFrame
columns = [f'PC{i+1}' for i in range(n_components)]  # نام‌گذاری ستون‌ها به صورت PC1, PC2, ...
df_reduced = pd.DataFrame(data_reduced, columns=columns)

# نمایش DataFrame کاهش‌یافته
df_reduced
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

# مدل درخت تصمیم
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# پیش‌بینی بر روی داده‌های آزمایشی
y_pred = model.predict(X_test)
```

<div id="sec_intro" style="direction: rtl; line-height: 300%;">
    <font face="XB Zar" size="4">
        <p></p>
        این کد از جستجوی شبکه‌ای (GridSearchCV) برای یافتن بهترین مقادیر هایپرپارامترهای مدل درخت تصمیم استفاده می‌کند. بهترین پارامترها و نمره عملکرد مدل بر اساس معیار F1 محاسبه و نمایش داده می‌شوند. همچنین مدت زمان اجرای جستجو نیز گزارش می‌شود.
    </font>
</div>

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [5, 10, 15, 20, 30],  # حداکثر عمق درخت تصمیم. None به معنای رشد درخت تا حد ممکن است.
    'min_samples_split': [2, 5, 10, 20],  # حداقل تعداد نمونه‌های لازم برای تقسیم یک گره. 
    'min_samples_leaf': [1, 2, 5, 10],    # حداقل تعداد نمونه‌های لازم برای برگ‌ها (leaf nodes).
    'criterion': ['gini', 'entropy'],  # معیار تقسیم گره‌ها: 
    'max_features': ['sqrt', 'log2', None],  # تعداد ویژگی‌ها برای انتخاب بهترین تقسیم:
                                             # 'sqrt' ریشه مربع تعداد ویژگی‌ها، 'log2' لگاریتم پایه 2، None تمام ویژگی‌ها.
    'splitter': ['best']  # روش انتخاب بهترین تقسیم: 'best' برای انتخاب بهترین تقسیم ممکن.
}


# شروع زمان سنجش
start_time = time.time()

# جستجوی شبکه‌ای برای بهترین پارامترها
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='f1', n_jobs=-1)

# آموزش مدل
grid_search.fit(X_train, y_train)

# بهترین مدل پیدا شده
best_model = grid_search.best_estimator_

# پایان زمان سنجش
end_time = time.time()

# نمایش هایپرپارامترهای بهینه‌شده
print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)

execution_time = end_time - start_time
print(f"Time taken for RandomizedSearchCV: {execution_time:.2f} seconds")
```

<div id="sec_intro" style="direction: rtl; line-height: 300%;">
    <font face="XB Zar" size="4">
        <p></p>
        این کد با استفاده از الگوریتم K-Means تعداد خوشه‌های بهینه (K) را تعیین می‌کند. ابتدا امتیاز سیلوئت برای مقادیر مختلف K (از 2 تا 10) محاسبه می‌شود. سپس مقدار K با بالاترین امتیاز سیلوئت به عنوان تعداد خوشه‌های بهینه انتخاب می‌شود. در نهایت، نمودار امتیازهای سیلوئت رسم شده و مقدار بهینه K با یک علامت مشخص می‌شود.
    </font>
</div>

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# داده‌ی خود را در X قرار دهید
X = df_cluster.values  # اگر دیتافریم دارید

K = range(2, 11)  # آزمون برای تعداد خوشه‌های مختلف
silhouette_scores = []

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# یافتن مقدار بیشینه و K مربوطه
max_score = max(silhouette_scores)  
optimal_k = K[silhouette_scores.index(max_score)]

print("Maximum Silhouette Score:", max_score)
print("Optimal K (by max Silhouette):", optimal_k)

# رسم نمودار سیلوئت
plt.plot(K, silhouette_scores, 'bo-', label='Silhouette Score')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for different K")
plt.scatter(optimal_k, max_score, c='red', s=200, label=f'Max (K={optimal_k})')
plt.legend()
plt.show()
```
